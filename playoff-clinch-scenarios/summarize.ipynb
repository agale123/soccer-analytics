```python
import pandas as pd
import itertools
```

```python
W = "(1, 0)"
D = "(0, 0)"
L = "(0, 1)"

OUTCOMES = [W, D, L]
```

```python
df = pd.read_csv("week_21_4.csv")
df["clinched"] = df["clinched"].str.strip("[]").str.split(",")
df["eliminated"] = df["eliminated"].str.strip("[]").str.split(",")
df["host"] = df["host"].str.strip("[]").str.split(",")
df["bye"] = df["bye"].str.strip("[]").str.split(",")

df
```

```python
TEAMS_CLINCHED = set(df["clinched"].sum()) - {""}
ALREADY_CLINCHED = {"SD", "POR"}

TEAMS_ELIMINATED = set(df["eliminated"].sum()) - {""}
ALREADY_ELIMINATED = {"KCC", "CHI"}

TEAMS_HOST = set(df["host"].sum()) - {""}
ALREADY_HOST = {"SD", "POR"}

TEAMS_BYE = set(df["bye"].sum()) - {""}
ALREADY_BYE = {"SD", "POR"}
```

```python
GAMES = list(filter(lambda x: "-" in x, df.columns))
GAMES
```

```python
def evaluate(teams, metric):
    for team in teams:
        print(team)

        # As we identify matching criteria, remove those rows
        subset = df.copy()

        # Store all the matches
        results = []

        # Check if all entries of a previous result are fully contained in values.
        def results_contains(values):
            for result in results:
                match = True
                for r in result:
                    if r not in values:
                        match = False
                if match:
                    return True
            return False

        # Evaluate single game criteria:
        for game in GAMES:
            for outcome in OUTCOMES:
                if (
                    df[df[game] == outcome][metric]
                    .str.contains(team, regex=False)
                    .all()
                ):
                    print(team, metric, "with:", game, outcome)
                    results.append([(game, outcome)])
                    subset = subset[~(subset[game] == outcome)]

        # Evaluate two game criteria
        pairs = []
        for game in GAMES:
            for outcome in subset[game].unique():
                pairs.append((game, outcome))

        for ((g1, o1), (g2, o2)) in itertools.combinations(pairs, 2):
            if g1 == g2:
                continue
            if results_contains([(g1, o1), (g2, o2)]):
                continue
            if (
                subset[(subset[g1] == o1) & (subset[g2] == o2)][metric]
                .str.contains(team, regex=False)
                .all()
            ):
                print(team, metric + " with:", g1, o1, g2, o2)
                subset = subset[~((subset[g1] == o1) & (subset[g2] == o2))]
                results.append([(g1, o1), (g2, o2)])

        # Evaluate three game criteria
        triples = []
        for game in GAMES:
            for outcome in subset[game].unique():
                triples.append((game, outcome))

        for ((g1, o1), (g2, o2), (g3, o3)) in itertools.combinations(triples, 3):
            if g1 == g2 or g2 == g3 or g3 == g1:
                continue
            if results_contains([(g1, o1), (g2, o2), (g3, o3)]):
                continue
            if (
                subset[(subset[g1] == o1) & (subset[g2] == o2) & (subset[g3] == o3)][
                    metric
                ]
                .str.contains(team, regex=False)
                .all()
            ):
                print(team, metric + " with:", g1, o1, g2, o2, g3, o3)
                subset = subset[
                    ~((subset[g1] == o1) & (subset[g2] == o2) & (subset[g3] == o3))
                ]
                results.append([(g1, o1), (g2, o2), (g3, o3)])

        # Evaluate four game criteria
        quads = []
        for game in GAMES:
            for outcome in subset[game].unique():
                quads.append((game, outcome))

        for ((g1, o1), (g2, o2), (g3, o3), (g4, o4)) in itertools.combinations(
            quads, 4
        ):
            if len([g1, g2, g3, g4]) != len(set([g1, g2, g3, g4])):
                continue
            if results_contains([(g1, o1), (g2, o2), (g3, o3), (g4, o4)]):
                continue
            if (
                subset[
                    (subset[g1] == o1)
                    & (subset[g2] == o2)
                    & (subset[g3] == o3)
                    & (subset[g4] == o4)
                ][metric]
                .str.contains(team, regex=False)
                .all()
            ):
                print(team, metric + " with:", g1, o1, g2, o2, g3, o3, g4, o4)
                subset = subset[
                    ~(
                        (subset[g1] == o1)
                        & (subset[g2] == o2)
                        & (subset[g3] == o3)
                        & (subset[g4] == o4)
                    )
                ]
                results.append([(g1, o1), (g2, o2), (g3, o3), (g4, o4)])
                
         # Evaluate five game criteria
        quints = []
        for game in GAMES:
            for outcome in subset[game].unique():
                quints.append((game, outcome))

        for ((g1, o1), (g2, o2), (g3, o3), (g4, o4), (g5, o5)) in itertools.combinations(
            quints, 5
        ):
            if len([g1, g2, g3, g4, g5]) != len(set([g1, g2, g3, g4, g5])):
                continue
            if results_contains([(g1, o1), (g2, o2), (g3, o3), (g4, o4), (g5, o5)]):
                continue
            if (
                subset[
                    (subset[g1] == o1)
                    & (subset[g2] == o2)
                    & (subset[g3] == o3)
                    & (subset[g4] == o4)
                    & (subset[g5] == o5)
                ][metric]
                .str.contains(team, regex=False)
                .all()
            ):
                print(team, metric + " with:", g1, o1, g2, o2, g3, o3, g4, o4, g5, o5)
                subset = subset[
                    ~(
                        (subset[g1] == o1)
                        & (subset[g2] == o2)
                        & (subset[g3] == o3)
                        & (subset[g4] == o4)
                        & (subset[g5] == o5)
                    )
                ]
                results.append([(g1, o1), (g2, o2), (g3, o3), (g4, o4), (g5, o5)])
                
        # Evaluate six game criteria
        sexs = []
        for game in GAMES:
            for outcome in subset[game].unique():
                sexs.append((game, outcome))

        for ((g1, o1), (g2, o2), (g3, o3), (g4, o4), (g5, o5), (g6, o6)) in itertools.combinations(
            sexs, 6
        ):
            if len([g1, g2, g3, g4, g5, g6]) != len(set([g1, g2, g3, g4, g5, g6])):
                continue
            if results_contains([(g1, o1), (g2, o2), (g3, o3), (g4, o4), (g5, o5), (g6, o6)]):
                continue
            if (
                subset[
                    (subset[g1] == o1)
                    & (subset[g2] == o2)
                    & (subset[g3] == o3)
                    & (subset[g4] == o4)
                    & (subset[g5] == o5)
                    & (subset[g6] == o6)
                ][metric]
                .str.contains(team, regex=False)
                .all()
            ):
                print(team, metric + " with:", g1, o1, g2, o2, g3, o3, g4, o4, g5, o5, g6, o6)
                subset = subset[
                    ~(
                        (subset[g1] == o1)
                        & (subset[g2] == o2)
                        & (subset[g3] == o3)
                        & (subset[g4] == o4)
                        & (subset[g5] == o5)
                        & (subset[g6] == o6)
                    )
                ]
                results.append([(g1, o1), (g2, o2), (g3, o3), (g4, o4), (g5, o5), (g6, o6)])


evaluate(TEAMS_CLINCHED - ALREADY_CLINCHED, "clinched")
```

```python
evaluate(TEAMS_ELIMINATED-ALREADY_ELIMINATED, "eliminated")
```
