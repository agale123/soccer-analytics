```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
from common import Fonts, Colors

BLUE = "#262262"
RED = "#be1e2d"
```

```python
years = pd.DataFrame(
    [
        ["<2", 36],
        ["2-10", 37],
        ["10-20", 23],
        ["20+", 7],
    ],
    columns=["label", "count"],
)
years
```

```python
level = pd.DataFrame(
    [
        ["Level 100", 17],
        ["Level 200", 26],
        ["Level 300", 21],
        ["Level 400", 1],
        ["Other", 40],
    ],
    columns=["label", "count"],
)
level
```

```python
pay= pd.DataFrame(
    [
        ["Paid", 53],
        ["Unpaid", 46],
    ],
    columns=["label", "count"],
)
pay
```

```python
access = pd.DataFrame(
    [
        ["Fields", 83 + 21],
        ["Strength & conditioning facilities", 26 + 21],
        ["Trainers for practices and games", 12 + 21],
        ["Trainers, but must pay", 30 + 21],
        ["Medical coverage", 16 + 21],
        ["Atheltic department logo", 54 + 21],
    ],
    columns=["label", "count"],
)
access
```

```python
# Setup
plt.rcParams["figure.figsize"] = (10, 8)
plt.figure(dpi=200)
fig, ax = plt.subplots()
fig.patch.set_facecolor("white")
ax.set_facecolor("white")

ax.bar(years["label"], years["count"], color=BLUE)
ax.set_xlabel("Years", size=18, color=Colors.BLACK, fontproperties=Fonts.BARLOW)
ax.set_ylabel("Count", size=18, color=Colors.BLACK, fontproperties=Fonts.BARLOW)

# Axes
ax.spines[["top", "right", "left"]].set_visible(False)
ax.tick_params(bottom=False, left=False)
ax.yaxis.grid(True, color=Colors.LIGHT_GRAY)
ax.set_axisbelow(True)
ax.tick_params(axis="both", which="major", labelsize=16, labelcolor=Colors.BLACK)
ax.set_ylim([0,43])

# Title
ax.set_title(
    "Years coaching with current team",
    fontproperties=Fonts.BARLOW_BOLD,
    size=32,
    x=0,
    ha="left",
    color=Colors.BLACK,
)

# Save image
fig.savefig("years.png", bbox_inches="tight", pad_inches=0.5, dpi=200)
```

```python
# Setup
plt.rcParams["figure.figsize"] = (10, 8)
plt.figure(dpi=200)
fig, ax = plt.subplots()
fig.patch.set_facecolor("white")
ax.set_facecolor("white")

ax.bar(level["label"], level["count"], color=BLUE)
ax.set_ylabel("Count", size=18, color=Colors.BLACK, fontproperties=Fonts.BARLOW)

# Axes
ax.spines[["top", "right", "left"]].set_visible(False)
ax.tick_params(bottom=False, left=False)
ax.yaxis.grid(True, color=Colors.LIGHT_GRAY)
ax.set_axisbelow(True)
ax.tick_params(axis="both", which="major", labelsize=16, labelcolor=Colors.BLACK)

# Title
ax.set_title(
    "Highest certification",
    fontproperties=Fonts.BARLOW_BOLD,
    size=32,
    x=0,
    ha="left",
    color=Colors.BLACK,
)

# Save image
fig.savefig("level.png", bbox_inches="tight", pad_inches=0.5, dpi=200)
```

```python
# Setup
plt.rcParams["figure.figsize"] = (10, 8)
plt.figure(dpi=200)
fig, ax = plt.subplots()
fig.patch.set_facecolor("white")
ax.set_facecolor("white")

ax.barh(access["label"], access["count"]/105*100, color=BLUE)

# Axes
ax.spines[["top", "right", "bottom"]].set_visible(False)
ax.tick_params(bottom=False, left=False)
ax.xaxis.grid(True, color=Colors.LIGHT_GRAY)
ax.set_axisbelow(True)
ax.tick_params(axis="both", which="major", labelsize=16, labelcolor=Colors.BLACK)
ax.xaxis.set_major_formatter(mtick.PercentFormatter())

# Title
ax.set_title(
    "Access to University resources",
    fontproperties=Fonts.BARLOW_BOLD,
    size=32,
    x=0,
    ha="left",
    color=Colors.BLACK,
)

# Save image
fig.savefig("access.png", bbox_inches="tight", pad_inches=0.5, dpi=200)
```

```python
# Setup
plt.rcParams["figure.figsize"] = (10, 8)
plt.figure(dpi=200)
fig, ax = plt.subplots()
fig.patch.set_facecolor("white")
ax.set_facecolor("white")

_, _, autotexts = ax.pie(
    pay["count"],
    labels=pay["label"],
    autopct="%1.0f%%",
    colors=[BLUE, RED],
    textprops={"fontsize": 18, "ha": "center"},
    labeldistance=1.3,
    startangle=90,
)
for autotext in autotexts:
    autotext.set_color("white")

# Title
ax.set_title(
    "Pay disparities",
    fontproperties=Fonts.BARLOW_BOLD,
    size=32,
    x=0,
    ha="left",
    color=Colors.BLACK,
)

# Save image
fig.savefig("pay.png", bbox_inches="tight", pad_inches=0.5, dpi=200)
```
