```python
from common import Data
import pandas as pd
```

```python
df = Data.get_nwsl_matches()
df["stadium"] = df["stadium"].apply(lambda x: x.replace(" (Neutral Site)", ""))
df["goals"] = (df["home_score"] + df["away_score"])
df
```

```python
df[(df["season"] != "2020 Challenge Cup") & ~df["is_playoffs"]].groupby(
    ["home", "stadium"]
).agg(count=("date", "count")).reset_index()
```

```python
df[
    (df["season"] != "2020 Challenge Cup") & ~df["is_playoffs"] & (df["home"] == "WAS")
].groupby(["home", "season", "stadium"]).agg(count=("date", "count")).reset_index()
```

```python
df[df["is_playoffs"]]
```

```python
df.groupby(["stadium"]).agg(
    count=("date", "count"),
    goals=("goals", "sum"),
    max_attendance=("attendance", "max"),
).sort_values("count", ascending=False).reset_index().to_csv(
    "stadiums.csv", index=False
)
```
