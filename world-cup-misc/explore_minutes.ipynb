```python
import pandas as pd
```

```python
year = "2023"
df = pd.read_csv("data/minutes_" + year + ".csv")
df
```

```python
quantiles = pd.DataFrame([], columns=["team", "25th", "50th", "75th"])

for team in df["team"].unique():
    subset = df[df["team"] == team]
    subset = subset.drop(columns=["team"])
    subset = subset.set_index("player")
    subset = subset.dropna(axis=1, how="all")
    subset = subset.clip(0, 90).astype(int)
    subset["total"] = subset.sum(axis=1) / len(subset.columns)
    quantiles.loc[len(quantiles.index)] = [
        team,
        subset["total"].quantile(0.25),
        subset["total"].quantile(0.5),
        subset["total"].quantile(0.75),
    ]
```
