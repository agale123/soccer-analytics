```python
import numpy as np
import pandas as pd
```

```python
year = "2015"
df = pd.read_csv("data/raw_minutes_" + year + ".csv")
df = df.rename(columns = {
    "Player_Name": "player",
    "Team": "team",
    "Matchday": "date",
    "Min": "minutes",
})
df
```

```python
result = pd.DataFrame(
    [],
    columns=[
        "team",
        "group_1",
        "group_2",
        "group_3",
        "round_of_16",
        "quarterfinals",
        "semifinals",
        "finals",
    ],
)
result.index.name = "player"


def get_i_row(df, player, i):
    subset = df[df["player"] == player].sort_values("date", ascending=True)
    team = subset.iloc[0]["team"]
    dates = df[df["team"] == team]["date"].unique()
    if i < len(dates):
        subset = subset[subset["date"] == dates[i]]
        if len(subset.index) < 1:
            return 0
        else:
            time = subset.iloc[0]["minutes"]
            return 0 if pd.isna(time) else time
    else:
        return None


for player in df["player"].unique():
    result.loc[player] = [df[df["player"] == player].iloc[0]["team"]] + [
        get_i_row(df, player, i) for i in range(0, 7)
    ]


result
```

```python
result.to_csv("data/minutes_" + year + ".csv")
```
