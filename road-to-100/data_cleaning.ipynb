```python
import nwslpy
import pandas as pd
import datetime
```

# Read 2016-now data from nwslR

```python
teams = nwslpy.load_teams()
teams
```

```python
matches = nwslpy.load_matches()
matches.head()
```

```python
merged = (
    matches.merge(
        teams[["team_abbreviation"]], left_on="home_team_id", right_on="team_id"
    )
    .rename(columns={"team_abbreviation": "home_team"})
    .merge(teams[["team_abbreviation"]], left_on="away_team_id", right_on="team_id")
    .rename(
        columns={
            "team_abbreviation": "away_team",
        }
    )
    .sort_values("kickoff")
)

merged = merged[
    [
        "kickoff",
        "season",
        "home_team",
        "away_team",
        "home_team_score",
        "away_team_score",
        "home_penalty_score",
        "away_penalty_score",
    ]
]
merged
```

```python
# Calculate whether each game is part of the playoffs
playoffs = {
    "2023 Challenge Cup": datetime.date(2023, 8, 7),
    "2023": datetime.date(2023, 10, 21),
    "2022 Challenge Cup": datetime.date(2022, 5, 3),
    "2022": datetime.date(2022, 10, 15),
    "2021 Challenge Cup": datetime.date(2021, 5, 7),
    "2021": datetime.date(2021, 11, 6),
    "2020 Fall Series": datetime.date(2020, 10, 18),
    "2020 Challenge Cup": datetime.date(2020, 7, 16),
    "2019": datetime.date(2019, 10, 19),
    "2018": datetime.date(2018, 9, 14),
    "2017": datetime.date(2017, 10, 6),
    "2016": datetime.date(2016, 9, 29),
}

merged["is_playoff"] = merged.apply(
    lambda row: row["kickoff"] > playoffs[row["season"]], axis=1
)
merged
```

# Read 2013-2015 data from fbref

```python
years = ["2013", "2014", "2015"]

fbref = pd.DataFrame([]);

for year in years:
    df = pd.read_csv("data/" + year + ".csv")
    teams = {
        "Dash": "HOU",
        "Thorns": "POR",
        "Spirit": "WAS",
        "Reign": "RGN",
        "Kansas City": "KC",
        "Sky Blue": "NJY",
        "Flash": "WNY",
        "Boston": "BOS",
        "Red Stars": "CHI",
    }
    df["home_team"] = df["Home"].apply(lambda x: teams[x])
    df["away_team"] = df["Away"].apply(lambda x: teams[x])
    df["kickoff"] = df.apply(
        lambda row: row["Date"] + " " + row["Time"].split(" ")[0], axis=1
    )
    df["kickoff"] = pd.to_datetime(df["kickoff"])
    df["home_team_score"] = df.apply(lambda row: int(row["Score"].split("–")[0]), axis=1)
    df["away_team_score"] = df.apply(lambda row: int(row["Score"].split("–")[1]), axis=1)
    df["home_penalty_score"] = 0
    df["away_penalty_score"] = 0
    df["season"] = year
    df["is_playoff"] = df["Round"] != "Regular Season"

    df = df[
        [
            "kickoff",
            "season",
            "home_team",
            "away_team",
            "home_team_score",
            "away_team_score",
            "home_penalty_score",
            "away_penalty_score",
            "is_playoff"
        ]
    ]
    
    fbref = fbref.append(df)
fbref
```

```python
# Write combined dataset to file
fbref.append(merged).to_csv("data/all_matches.csv", index=False)
```
