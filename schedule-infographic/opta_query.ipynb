```python
import requests
import pandas as pd
from datetime import datetime
```

```python
url = 'https://api.performfeeds.com/soccerdata/match/1u2vu9dy9um7r1galdsvmxmdkm?_rt=c&_fmt=json&_lcl=en-gb&lineups=no&live=yes&comp=djkvh1luqdkdixpoe653lzixg%2C85qdzpr6i56z8tjex9d5mu5pm%2C135j6qi3ap8i3vc9hwdo4fjtg&_pgSz=250&_pgNm=1&mt.mDt=%5B2026-01-16T08%3A00%3A00.000Z+TO+2026-12-01T06%3A59%3A59.999Z%5D'

# Define the headers as a dictionary
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/143.0.0.0 Safari/537.36',
    'Accept': 'application/json',
    'Referer': 'https://live.opta.statsperform.com/',
}

response = requests.get(url, headers=headers)
```

```python
raw = response.json()
```

```python
def format_match(match):
    date=match['matchInfo']['localDate']
    time=datetime.strptime(match['matchInfo']['localTime'], '%H:%M:%S')
    time=time.strftime('%I:%M %p')
    home=match['matchInfo']['contestant'][0]['code']
    away=match['matchInfo']['contestant'][1]['code']
    
    return [date, time, home, away]
```

```python
df = pd.DataFrame([], columns=["date", "time", "home", "away"])
for match in raw['match']:
    df.loc[len(df)] =format_match(match)

df
```

```python
df = df.sort_values('date')
df
```

```python
df = df.replace({'SEA': 'RGN', 'KC': 'KCC'})
```

```python
df.to_csv("data/2026_schedule.csv", index=False)
```
