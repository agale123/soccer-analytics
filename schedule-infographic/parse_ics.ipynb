```python
import pandas as pd
from icalendar import Calendar, Event
from zoneinfo import ZoneInfo
from common import NWSL
```

```python
df = pd.DataFrame([], columns=["date", "time", "home", "away"])
df
```

```python
et = ZoneInfo("America/New_York")
ct = ZoneInfo("America/Chicago")
mt = ZoneInfo("America/Denver")
pt = ZoneInfo("America/Los_Angeles")

tzs = {
    "GFC": et,
    "POR": pt,
    "LA": pt,
    "CHI": ct,
    "WAS": et,
    "SD": pt,
    "NC": et,
    "RGN": pt,
    "LOU": et,
    "HOU": ct,
    "ORL": et,
    "KCC": ct,
    "UTA": mt,
    "BAY": pt,
    "DEN": mt,
    "BOS": et,
}
```

```python
f = open('data/NWSL.ics','rb')
cal = Calendar.from_ical(f.read())
for component in cal.walk():
    if component.name == "VEVENT":
        try:
            summary = component.get('summary')[3:]
            home = NWSL.map_team(summary.split(" vs ")[0])
            away = NWSL.map_team(summary.split(" vs ")[1])
            date = component.get('dtstart').dt.astimezone(tzs[home])
            df.loc[len(df)] =[date.date(), date.time().strftime("%-I:%M %p"),home, away]
        except:
            print('pass')
f.close()

df
```

```python
decision_day = pd.DataFrame(
    [
        ["2026-11-01", "", "LA", "BOS"],
        ["2026-11-01", "", "BAY", "SD"],
        ["2026-11-01", "", "GFC", "UTA"],
        ["2026-11-01", "", "HOU", "POR"],
        ["2026-11-01", "", "NC", "DEN"],
        ["2026-11-01", "", "LOU", "KCC"],
        ["2026-11-01", "", "RGN", "ORL"],
        ["2026-11-01", "", "WAS", "CHI"],
    ],
    columns=["date", "time", "home", "away"],
)

df = pd.concat([df, decision_day])
```
