```python
from common import Data
```

```python
df = Data.get_nwsl_player_stats(include_playoffs=False, include_cups=False, season="2022")
df
```

```python
df = df[(df["team"] == "LA") | (df["team"] == "SD")]
df
```

```python
df = df.groupby(["player", "team"]).agg(
    minutes=("mins_played", "sum")
).reset_index().sort_values(["team", "player"]).reset_index(drop=True)
df
```

```python
df["acquired"] = ""
df
```

```python
df.to_csv("data/LA_SD_expansion.csv", index=False)
```
