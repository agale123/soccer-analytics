```python
import nwslpy
import pandas as pd
```

```python
uswnt = pd.read_csv("2023-callups.csv")
uswnt
```

```python
teams = nwslpy.load_teams()
teams = teams[teams["last_season"] >= 2023]
teams
```

```python
players = nwslpy.load_players()
players["name"] = players["player_short_first_name"] + " " + players["player_short_last_name"]
players = players[["name"]]
players
```

```python
df = pd.concat([nwslpy.load_player_season_stats(t, "2023") for t in teams["team_abbreviation"]])
df
```

```python
df = df.set_index("player_id").join(players)
df = df[["name", "minutes"]]
df
```

```python
uswnt = uswnt.rename(columns={"minutes": "uswnt_minutes"})
df = df.rename(columns={"minutes": "nwsl_minutes"})
```

```python
df = uswnt.set_index("name").join(df.set_index("name"))
```

```python
df = df.fillna(0)
df["nwsl_minutes"] = df["nwsl_minutes"].astype(int)
df
```

```python
df.to_csv("joined.csv")
```
