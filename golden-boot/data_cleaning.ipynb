```python
import pandas as pd
import numpy as np
from common import Data
from datetime import datetime, timedelta
```

```python
year = "2019"
```

```python
df = Data.get_nwsl_player_stats(season=year)
df = df[df["date"] < "2019-10-19"]
df
```

```python
# Get range of dates
date_list = pd.date_range(df["date"].min(), df["date"].max(), freq='D')
date_list
```

```python
# Get all goal scorers
scorers = df.groupby("player").agg({"goals": "sum"})
scorers = scorers[scorers["goals"] > 0]
scorers = scorers.sort_values(["goals"], ascending=False)
scorers
```

```python
# Create empty table
res = pd.DataFrame([], columns=date_list, index=scorers.index)
res
```

```python
# Fill in data
for d in date_list:
    today = str(d.date())
    yesterday = str((d - timedelta(days=1)).date())
    if today == str(df["date"].min()):
        res[today] = 0
    else:
        res[today] = res[yesterday]
        
    for i, row in df[df["date"] == today].iterrows():
        if row["player"] in scorers.index and not np.isnan(row["goals"]):
            res.loc[row["player"], d] = res.loc[row["player"], d] + row["goals"]
        
res
    
```

```python
res.to_csv("golden_boot_" + year + ".csv")
```
