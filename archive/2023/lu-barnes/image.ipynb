```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from common import Images, Fonts, Colors, add_twitter_handle, add_source
```

```python
def plot_image(title, subtitle, source, output):
    # Setup
    fig, ax = plt.subplots()
    plt.rcParams["figure.figsize"] = (14, 9)
    plt.figure(dpi=200)
    fig.patch.set_facecolor(Colors.WHITE)
    ax.set_facecolor(Colors.WHITE)

    # Plot image
    img = mpimg.imread(source)
    ax.imshow(img)
    ax.axis("off")

    # Title
    ax.set_title(
        title,
        fontproperties=Fonts.BARLOW_BOLD,
        size=28,
        x=0.0,
        y=1.12,
        ha="left",
        color=Colors.BLACK,
    )

    # Subtitle
    ax.text(
        0.0,
        1.03,
        subtitle,
        fontproperties=Fonts.BARLOW,
        size=16,
        ha="left",
        transform=ax.transAxes,
    )

    add_twitter_handle(fig, x=0.68, y=0.1, fs=14, w=0.028)
    add_source(fig, "nwslR/nwslpy & FBRef", x=0.24, y=0.1, fs=14)

    # Save image
    fig.savefig(output, bbox_inches="tight", pad_inches=0.5, dpi=200)
```

```python
plot_image(
    "Lauren Barnes: Road to 200",
    "Each hexagon represents a Reign game with Lauren Barnes' wins in blue, draws\nin gold, and losses in red.",
    "games.png",
    "1_games.png",
)
```

```python
plot_image(
    "Lauren Barnes: Road to 200",
    "Hexagons are scaled to represent minutes played and shaded to represent\nposition (left versus center back). Solid outlines represent a shutout by the Reign.",
    "minutes.png",
    "2_minutes.png",
)
```

```python
plot_image(
    "Lauren Barnes: Road to 200",
    "Assists are marked with a three-petal flower and her two goals are marked by a\nsix-petal flower.",
    "goals_assists.png",
    "3_goals_assists.png",
)
```
