```python
from itscalledsoccer.client import AmericanSoccerAnalysis
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from common import Fonts, Images, NWSL
```

```python
asa_client = AmericanSoccerAnalysis()
```

```python
teams = asa_client.get_teams(leagues="nwsl")
teams
```

```python
games = asa_client.get_games(leagues="nwsl")
games
```

```python
xgoals = asa_client.get_team_xgoals(leagues="nwsl", split_by_games=True)
xgoals
```

```python
# Join with the teams
xgoals = xgoals.set_index("team_id").join(teams.set_index("team_id")).reset_index()

# Join with the games
xgoals = xgoals.set_index("game_id").join(games.set_index("game_id")).reset_index()

xgoals
```

```python
xgoals = xgoals[
    [
        "team_name",
        "team_abbreviation",
        "season_name",
        "date_time_utc",
        "goals_for",
        "goals_against",
        "xgoals_for",
        "xgoals_against",
        "knockout_game",
    ]
]
xgoals
```

```python
xgoals.to_csv("xgoals.csv", index=False)
```
