```python
import requests
import json
import pandas as pd
from common import NWSL
```

```python
# Mar 31
week = 3
year = 2025
```

```python
data = requests.get("https://api-sdp.nwslsoccer.com/v1/nwsl/football/seasons/nwsl::Football_Season::fad050beee834db88fa9f2eb28ce5a5c/standings?locale=en-US&orderBy=rank&direction=asc").json()
data
```

```python
def convert_team(t):
    stats = {}
    for stat in t["stats"]:
        stats[stat["statsId"]] = stat["statsValue"]
    return [
        t["acronymName"],
        stats["rank"],
        stats["points"],
        stats["matches-played"],
        stats["win"],
        stats["lose"],
        stats["draw"],
        stats["goals-for"],
        stats["goals-against"],
        stats["goal-difference"],
    ]


df = pd.DataFrame(
    map(convert_team, data["teams"]),
    columns=[
        "team",
        "rank",
        "PTS",
        "MP",
        "W",
        "L",
        "D",
        "GF",
        "GA",
        "GD"
    ],
)
df = df.replace("KC", "KCC")
df = df.replace("NJY", "NJNY")
df = df.replace("GFC", "NJNY")
df = df.replace("SEA", "RGN")
df
```

```python
df.to_csv("data/" + str(year) + "/" + str(week) + ".csv", index=False)
```
