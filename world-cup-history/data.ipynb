```python
import pandas as pd
```

```python
results = pd.read_csv("results.csv")
results
```

```python
feds = pd.read_csv("federations.csv")
feds
```

```python
hosts = pd.read_csv("host.csv")
hosts
```

```python
df = (
    results.pivot(index="team", columns="year", values="result")
    .rename_axis(None, axis=1)
)
df
```

```python
df = (
    df.join(feds.set_index("team"))
    .sort_index(axis=0)
    .sort_values(["federation"], kind="stable")
)
df
```

```python
# Add row for index
df = df.reset_index()
df.loc[-1] = ["Legend"] + (["X"] * (len(df.columns)-2)) + ["Legend"]
df.index = df.index + 1  # shifting index
df.sort_index(inplace=True) 
df
```

```python
df.to_csv("merged.csv", index=False)
```
