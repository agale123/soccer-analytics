```python
from common import Data
import pandas as pd
```

```python
df = Data.get_nwsl_matches(season=2025)
df
```

```python
df["home_win"] = df["home_score"] > df["away_score"]
df["home_loss"] = df["home_score"] < df["away_score"]
df["home_draw"] = df["home_score"] == df["away_score"]
df["away_win"] = df["home_score"] < df["away_score"]
df["away_loss"] = df["home_score"] > df["away_score"]
df["away_draw"] = df["home_score"] == df["away_score"]
df
```

```python
df_home = (
    df.groupby("home")
    .agg({"home_win": "sum", "home_loss": "sum", "home_draw": "sum"})
    .reset_index()
    .rename(columns={"home": "team"})
)
df_home
```

```python
df_away = (
    df.groupby("away")
    .agg({"away_win": "sum", "away_loss": "sum", "away_draw": "sum"})
    .reset_index()
    .rename(columns={"away": "team"})
)
df_away
```

```python
df = pd.merge(df_home, df_away, on='team')

df["home_w_l_d"] = list(zip(df["home_win"], df["home_loss"], df["home_draw"]))
df["away_w_l_d"] = list(zip(df["away_win"], df["away_loss"], df["away_draw"]))
df["home_points"] = df["home_win"] * 3 + df["home_draw"] * 1
df["home_mp"] = df["home_win"] + df["home_loss"] + df["home_draw"]
df["home_points_max"] = 3 * df["home_mp"]
df["away_points"] = df["away_win"] * 3 + df["away_draw"] * 1
df["away_mp"] = df["away_win"] + df["away_loss"] + df["away_draw"]
df["away_points_max"] = 3 * df["away_mp"]

df = df.drop(columns=["home_win", "home_loss", "home_draw", "home_mp", "away_win", "away_loss", "away_draw", "away_mp"])

df["diff"] = df["home_points"] - df["away_points"]

df
```
