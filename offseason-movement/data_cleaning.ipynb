```python
import pandas as pd
```

```python
league = "nwsl"
```

```python
# Read the rosters at the end of 2023
old = pd.read_csv(league + "/2023_end.csv")
old
```

```python
# Read the rosters at the beginning of 2024
new = pd.read_csv(league + "/2024_start.csv")
new
```

```python
# Join the rosters
df = (
    old.set_index("player")
    .join(new.set_index("player"), how="outer", lsuffix="_old", rsuffix="_new")
    .reset_index()
)
df
```

```python
# Find all the duplicate entries
pd.concat(g for _, g in df.groupby("player") if len(g) > 1)
```

```python
# Check that all newly signed players are correct
signed = df[df["team_old"].isna()].sort_values("player")
signed.to_csv(league + "/signed.csv", index=False)
signed
```

```python
# Check that all dropped players are correct
dropped = df[df["team_new"].isna()].sort_values("player")
dropped.to_csv(league + "/dropped.csv", index=False)
dropped
```

```python
# Write output to file
df.to_csv(league + "/2023_2024_movement.csv", index=False)
```
