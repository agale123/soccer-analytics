```python
import pandas as pd
```

```python
league = "nwsl"
```

```python
# Read the rosters at the end of 2022
old = pd.read_csv(league + "/2022_end.csv")
old
```

```python
# Read the rosters at the beginning of 2023
new = pd.read_csv(league + "/2023_start.csv")
new
```

```python
# Join the rosters
df = (
    old.set_index("player")
    .join(new.set_index("player"), how="outer", lsuffix="_old", rsuffix="_new")
    .reset_index()
)
df
```

```python
# Check that all newly signed players are correct
signed = df[df["team_old"].isna()].sort_values("team_new")
signed.to_csv("signed.csv", index=False)
signed
```

```python
# Check that all dropped players are correct
dropped = df[df["team_new"].isna()].sort_values("team_old")
dropped.to_csv(league + "/dropped.csv", index=False)
dropped
```

```python
# Write output to file
df.to_csv(league + "/2022_2023_movement.csv", index=False)
```
