```python
import nwslpy
import pandas as pd
import datetime
```

# Read 2016-now data from nwslR

```python
players = nwslpy.load_players()
player_id = players[players["player_last_name"] == "Barnes"].index[0]
player_id
```

```python
teams = nwslpy.load_teams()
team_id = teams[teams["team_abbreviation"] == "RGN"].index[0]
team_id
```

```python

stats = nwslpy.load_player_season_stats("RGN", "2016")
stats[stats["player_id"] == player_id]
stats = stats[["season"]]
```

```python
# Get all matches
matches = nwslpy.load_matches()
matches
```

```python
# Filter to Reign matches
matches = matches[(matches["home_team_id"] == team_id) | (matches["away_team_id"] == team_id)]

# Filter to regular season
matches = matches[matches["season"].isin(map(str, range(2016, 2024)))]

# Calculate whether each game is part of the playoffs
playoffs = {
    "2023 Challenge Cup": datetime.date(2023, 8, 7),
    "2023": datetime.date(2023, 10, 21),
    "2022 Challenge Cup": datetime.date(2022, 5, 3),
    "2022": datetime.date(2022, 10, 15),
    "2021 Challenge Cup": datetime.date(2021, 5, 7),
    "2021": datetime.date(2021, 11, 6),
    "2020 Fall Series": datetime.date(2020, 10, 18),
    "2020 Challenge Cup": datetime.date(2020, 7, 16),
    "2019": datetime.date(2019, 10, 19),
    "2018": datetime.date(2018, 9, 14),
    "2017": datetime.date(2017, 10, 6),
    "2016": datetime.date(2016, 9, 29),
}
matches["is_playoff"] = matches.apply(
    lambda row: row["kickoff"] > pd.Timestamp(playoffs[row["season"]]), axis=1
)
matches = matches[~matches["is_playoff"]]
```

```python
df = pd.DataFrame(
    [],
    columns=[
        "date",
        "season",
        "opponent",
        "goals_for",
        "goals_against",
        "goals",
        "assists",
        "minutes",
        "yellow_cards",
        "red_cards",
    ],
)
```

```python
dates = list(df["date"].apply(lambda x: x.strftime("%Y-%m-%d")))

for i, row in matches.iterrows():
    if row["kickoff"].strftime("%Y-%m-%d") in dates:
        continue
    elif row["kickoff"] > datetime.datetime.now():
        continue
    elif i == "houston-dash-vs-seatle-reign-fc-2017-09-03":
        df = df.append(
            {
                "date": row["kickoff"].date(),
                "season": row["season"],
                "opponent": "HOU",
                "goals_for": 1,
                "goals_against": 0,
                "goals": 0,
                "assists": 0,
                "minutes": 90,
                "yellow_cards": 0,
                "red_cards": 0,
            }, ignore_index=True
        )
        continue
    elif i == "ol-reign-vs-chicago-red-stars-2021-06-22":
        continue
    elif i == "chicago-red-stars-vs-ol-reign-2021-07-18":
        df = df.append(
            {
                "date": row["kickoff"].date(),
                "season": row["season"],
                "opponent": "CHI",
                "goals_for": 1,
                "goals_against": 3,
                "goals": 0,
                "assists": 0,
                "minutes": 90,
                "yellow_cards": 0,
                "red_cards": 0,
            }, ignore_index=True
        )
        continue
    elif i == "washington-spirit-vs-ol-reign-2021-09-12":
        continue
    elif i == "ol-reign-vs-chicago-red-stars-2021-10-10":
        df = df.append(
            {
                "date": row["kickoff"].date(),
                "season": row["season"],
                "opponent": "CHI",
                "goals_for": 3,
                "goals_against": 2,
                "goals": 0,
                "assists": 0,
                "minutes": 90,
                "yellow_cards": 0,
                "red_cards": 0,
            }, ignore_index=True
        )
        continue
    elif i == "ol-reign-vs-kansas-city-current-2022-05-25":
        df = df.append(
            {
                "date": row["kickoff"].date(),
                "season": row["season"],
                "opponent": "KCC",
                "goals_for": 1,
                "goals_against": 0,
                "goals": 0,
                "assists": 0,
                "minutes": 90,
                "yellow_cards": 0,
                "red_cards": 0,
            }, ignore_index=True
        )
        continue
    elif i == "washington-spirit-vs-ol-reign-2023-03-26":
        df = df.append(
            {
                "date": row["kickoff"].date(),
                "season": row["season"],
                "opponent": "WAS",
                "goals_for": 0,
                "goals_against": 1,
                "goals": 0,
                "assists": 0,
                "minutes": 90,
                "yellow_cards": 0,
                "red_cards": 0,
            }, ignore_index=True
        )
        continue
    elif i == "nj-ny-gotham-fc-vs-ol-reign-2023-04-01":
        df = df.append(
            {
                "date": row["kickoff"].date(),
                "season": row["season"],
                "opponent": "NJY",
                "goals_for": 2,
                "goals_against": 0,
                "goals": 0,
                "assists": 0,
                "minutes": 90,
                "yellow_cards": 0,
                "red_cards": 0,
            }, ignore_index=True
        )
        continue
    try:
        stats = nwslpy.load_player_match_stats(i)
    except:
        print(i)
        break
    if "player_id" not in stats:
        print(i)
        continue
    stats = stats[stats["player_id"] == player_id]
    if len(stats.index) < 1:
        continue
    stats = stats.iloc[0]

    if row["home_team_id"] == team_id:
        opponent = row["away_team_id"]
        goals_for = row["home_team_score"]
        goals_against = row["away_team_score"]
    else:
        opponent = row["home_team_id"]
        goals_for = row["away_team_score"]
        goals_against = row["home_team_score"]
    if "minutes" in stats and stats["minutes"] > 0:
        df = df.append(
            {
                "date": row["kickoff"].date(),
                "season": row["season"],
                "opponent": teams.loc[opponent]["team_abbreviation"],
                "goals_for": goals_for,
                "goals_against": goals_against,
                "goals": stats["goals"] if "goals" in stats else 0,
                "assists": stats["assists_total"] if "assists_total" in stats else None,
                "minutes": stats["minutes"],
                "yellow_cards": stats["yellow_cards"] if "yellow_cards" in stats else 0,
                "red_cards": stats["red_cards"] if "red_cards" in stats else 0,
            }, ignore_index=True
        )
df
```

# Read 2013-2015 data from fbref

```python
years = ["2013", "2014", "2015"]

fbref = pd.DataFrame([]);

for year in years:
    df = pd.read_csv("data/" + year + ".csv")
    teams = {
        "Dash": "HOU",
        "Thorns": "POR",
        "Spirit": "WAS",
        "Reign": "RGN",
        "Kansas City": "KC",
        "Sky Blue": "NJY",
        "Flash": "WNY",
        "Boston": "BOS",
        "Red Stars": "CHI",
    }
    df["opponent"] = df["Opponent"].apply(lambda x: teams[x])
    df["date"] = df["Date"]
    df["goals_for"] = df.apply(lambda row: int(row["Result"][2:].split("â€“")[0]), axis=1)
    df["goals_against"] = df.apply(lambda row: int(row["Result"][2:].split("â€“")[1]), axis=1)
    df["season"] = year
    df["is_playoff"] = df["Round"] != "Regular Season"
    df["goals"] = df["Gls"]
    df["assists"] = df["Ast"]
    df["minutes"] = df["Min"]
    df["yellow_cards"] = df["CrdY"]
    df["red_cards"] = df["CrdR"]

    df = df[
        [
            "date",
            "season",
            "opponent",
            "goals_for",
            "goals_against",
            "goals",
            "assists",
            "minutes",
            "yellow_cards",
            "red_cards",
            "is_playoff"
        ]
    ]
    
    fbref = fbref.append(df)
    
# Remove playoff games and column
fbref = fbref[~fbref["is_playoff"]]
fbref = fbref.drop(columns=["is_playoff"])

fbref
```

```python
# Convert to datetimes
fbref["date"] = pd.to_datetime(fbref["date"])
df["date"] = pd.to_datetime(df["date"])

df = fbref.append(df).sort_values("date")
```

```python
# Cleaning duplicates
df["date"] = pd.to_datetime(df["date"])
df = df.drop_duplicates(subset="date", keep="first")
df = df.reset_index(drop=True)
df = df[df["is_playoff"] != True]
df = df.drop(columns=["is_playoff"])
df
```

```python
df.to_csv("data/all_matches.csv", index=False)
df
```
