```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from common import Images, Fonts, Colors, add_twitter_handle, add_source
```

```python
# Setup
fig, ax = plt.subplots()
plt.rcParams["figure.figsize"] = (14, 9)
plt.figure(dpi=200)
fig.patch.set_facecolor(Colors.WHITE)
ax.set_facecolor(Colors.WHITE)

# Plot image
img = mpimg.imread("raw.png")
ax.imshow(img)
ax.axis("off")

# Title
ax.set_title(
    "Lauren Barnes: Road to 200",
    fontproperties=Fonts.BARLOW_BOLD,
    size=32,
    x=0.0,
    y=1.12,
    ha="left",
    color=Colors.BLACK,
)

# Subtitle
ax.text(
    0.0,
    1.03,
    "Barnes is the first NWSL player to hit 200 regular season appearances. As a longtime captain and\nstalwart defender, she has been instrumental to the Reign's success.",
    fontproperties=Fonts.BARLOW,
    size=18,
    ha="left",
    transform=ax.transAxes,
)

add_twitter_handle(fig, x=0.79, y=0.12, fs=16, w=0.028)
add_source(fig, "nwslR/nwslpy & FBRef", x=0.13, y=0.12, fs=16)

# Save image
fig.savefig("barnes.png", bbox_inches="tight", pad_inches=0.5, dpi=200)
```
