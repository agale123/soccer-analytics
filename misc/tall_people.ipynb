```python
from common import Data
import pandas as pd
```

```python
df = Data.get_nwsl_player_stats(
    include_playoffs=True,
    include_early=True,
    include_cups=True,
)
df
```

```python
players = Data.get_nwsl_players()
players["is_6_foot"] = players["height"].str.contains("6'")
players["is_6_foot_not_gk"] = players["height"].str.contains("6'") & (players["position"] != "Goalkeeper")
players
```

```python
players[players["is_6_foot_not_gk"] == True]
```

```python
agg = df.set_index("player").join(players.set_index("name")[["is_6_foot", "is_6_foot_not_gk"]]).groupby(
    ["date", "team"]
).agg(tall=("is_6_foot_not_gk", "sum")).reset_index().sort_values(["tall"], ascending=False)
agg.iloc[0:20]
```
