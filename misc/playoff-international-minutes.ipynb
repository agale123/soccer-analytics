```python
from common import NWSL, Data, Colors, Fonts, Images, add_twitter_handle
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame(
    [["SD", 753], ["POR", 539], ["RGN", 977], ["NJNY", 348]],
    columns=["team", "minutes"],
)
df
```

```python
plt.clf()
plt.rcParams["figure.figsize"] = (12, 10)
plt.figure(dpi=200, facecolor=Colors.WHITE)
fig, ax = plt.subplots()
fig.patch.set_facecolor(Colors.WHITE)
ax.set_facecolor(Colors.WHITE)

ax.bar(df["team"], df["minutes"])

ax.set_ylabel("Minutes", fontproperties=Fonts.BARLOW, size=22)
ax.set_xticklabels(list(map(lambda x: NWSL.name(x), df["team"])))
ax.tick_params(axis='both', which='major', labelsize=16,length=0)
ax.spines[['top', 'right']].set_visible(False)

# Title
ax.set_title(
    "International Break Minutes",
    fontproperties=Fonts.BARLOW_BOLD,
    size=36,
    ha="left",
    x=0,
    y=1.1,
)

# Subtitle
ax.text(
    -0.59,
    1050,
    "OL Reign players were heavily utilized during the international break ahead\nof the semis, while Gotham players had more rest.",
    fontproperties=Fonts.BARLOW,
    size=20,
    ha="left",
)


add_twitter_handle(fig, x=0.8, y=0.04, fs=14, w=0.025)

# Save image
plt.savefig(
    "international_minutes.png",
    bbox_inches="tight",
    pad_inches=0.5,
    transparent=True,
)
```
