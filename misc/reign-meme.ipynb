```python
from common import NWSL, Data, Colors, Fonts, Images, add_twitter_handle
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame(
    [
        ["OL Reign", 0.95, "#e62929"],
        ["OL Reign, but in blue", 0.05, "#003692"],
    ],
    columns=["label", "value", "color"],
)
df
```

```python
# Setup
plt.clf()
plt.rcParams["figure.figsize"] = (12, 10)
plt.figure(dpi=400)
fig, ax = plt.subplots()

# PLot data
plt.pie(df["value"], colors=df["color"], startangle=90)

# Legend
plt.rcParams["legend.handlelength"] = 1 * 1
plt.rcParams["legend.handleheight"] = 1 * 1.2
plt.legend(
    labels=df["label"],
    frameon=False, prop={"size": 50}, bbox_to_anchor=(0.9, 0.9)
)

# Title
ax.set_title(
    "NWSL Championship Odds",
    fontproperties=Fonts.BARLOW_BOLD,
    size=80,
    ha="left",
    x=0,
    y=1.1,
)

# Add Twitter handle
add_twitter_handle(fig, x=1.26, y=0, fs=40, w=0.05)

# Save image
plt.savefig(
    "images/reign_meme.png",
    bbox_inches="tight",
    pad_inches=0.5,
    dpi=600,
    transparent=True,
)
```
