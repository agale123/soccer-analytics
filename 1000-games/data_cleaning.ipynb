```python
import nwslpy
import pandas as pd
from datetime import datetime
```

# Data from nwslR and FBRef

Data is copied over from `road-to-100/data/all_matches.csv`

```python
df = pd.read_csv("data/matches.csv")
df["kickoff"] = pd.to_datetime(df["kickoff"])
df
```

```python
# Filter to the regular season
df = df[df["season"].isin(list(map(str, range(2013, 2024))))]

# Remove playoff games
df = df[~df["is_playoff"]]

# Remove future games
df = df[df["kickoff"] < datetime.now()]

# Drop kickoff time
df["kickoff"] = df["kickoff"].dt.date

# Shorten column names
df = df.rename(
    columns={
        "kickoff": "date",
        "home_team": "home",
        "away_team": "away",
        "home_team_score": "home_score",
        "away_team_score": "away_score",
    }
)

# Remove unnecessary columns
df = df[["date", "season", "home", "away", "home_score", "away_score"]]

# Reset index
df = df.reset_index(drop=True)

df
```

```python
df.to_csv("data/cleaned.csv", index=False)
```
