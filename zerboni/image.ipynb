```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from common import Images, Fonts, Colors, add_twitter_handle, add_source
```

```python
def plot_image(title, subtitle, source, output):
    # Setup
    fig, ax = plt.subplots()
    plt.rcParams["figure.figsize"] = (14, 9)
    plt.figure(dpi=200)
    fig.patch.set_facecolor(Colors.WHITE)
    ax.set_facecolor(Colors.WHITE)

    # Plot image
    img = mpimg.imread(source)
    ax.imshow(img)
    ax.axis("off")

    # Title
    ax.set_title(
        title,
        fontproperties=Fonts.BARLOW_BOLD,
        size=28,
        x=0.0,
        y=1.12,
        ha="left",
        color=Colors.BLACK,
    )

    # Subtitle
    ax.text(
        0.0,
        1.03,
        subtitle,
        fontproperties=Fonts.BARLOW,
        size=16,
        ha="left",
        transform=ax.transAxes,
    )

    add_twitter_handle(fig, x=0.78, y=0.1, fs=14, w=0.028)
    add_source(fig, "nwslR/nwslpy & FBRef", x=0.15, y=0.1, fs=14)

    # Save image
    fig.savefig(output, bbox_inches="tight", pad_inches=0.5, dpi=200)
```

```python
plot_image(
    "McCall Zerboni: Road to 200",
    "Zerboni is the 2nd player to reach the 200 regular season match milestone. She has 16 regular season goals,\nand over 15,000 minutes playing for 5 different NWSL teams.",
    "raw.png",
    "zerboni.png",
)
```
