```python
import pandas as pd
```

```python
df = pd.read_csv("2019.csv")
df
```

```python
df = df[df["club"].apply(lambda x: "United States" in x)]
df
```

```python
df["club"] = df["club"].apply(lambda x: x.replace("United States ", ""))
df
```

```python
college = ['Southeastern Fire',
       'UCLA Bruins', 
       'UNC Wilmington Seahawks', 'Memphis Tigers',
       'Rutgers Scarlet Knights', 'UCF Knights', 'Montverde Academy',
       'Tennessee Volunteers', 'West Florida Argonauts',
       'United Soccer Alliance', 'Seattle Sounders', 
       'Florida Gators', 'Texas Longhorns',
       'Florida State Seminoles', 'California Golden Bears',
       'Kennesaw State Owls']
df = df[df["club"].apply(lambda x: x not in college)]
df
```

```python
df.to_csv("nwsl_2019.csv")
```
