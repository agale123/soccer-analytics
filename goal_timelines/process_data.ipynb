```python
import pandas as pd
```

```python
df = pd.read_csv("data/nwsl_goals_2023.csv")
df
```

```python
cols = {
    "0-15": (1, 0, 15),
    "15-30": (1, 15, 30),
    "30-45": (1, 30, 45),
    "45+": (1, 45, 90),
    "45-60": (2, 45, 60),
    "60-75": (2, 60, 75), 
    "75-90": (2, 75, 90), 
    "90+": (2, 90, 135)
}

for col in cols:
    half, lower, upper = cols[col]
    df[col] = (df["Event_Half"] == half) & (df["Event_Time"] >= lower) & (df["Event_Time"] < upper)
    
df = df.drop(columns=["Event_Time", "Event_Half"])
    
df
```

```python
scored = df.groupby("Team").sum()
scored.to_csv("data/goals_scored_2023.csv")
scored
```

```python
scored["1"] = scored["0-15"] + scored["15-30"] + scored["30-45"] + scored["45+"] 
scored["2"] = scored["45-60"] + scored["60-75"] + scored["75-90"] + scored["90+"] 
scored["ratio"] = scored["1"]/ (scored["1"] + scored["2"])
scored.sort_values(["ratio"])
```

```python
conceded = df.groupby("Opponent").sum()
conceded.to_csv("data/goals_conceded_2023.csv")
conceded
```
