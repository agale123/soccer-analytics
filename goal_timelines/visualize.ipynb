```python
import math
import pandas as pd
import matplotlib.pyplot as plt
from common import Fonts, Images, NWSL, Colors, add_twitter_handle, add_source
```

```python
scored = pd.read_csv("data/goals_scored_2023.csv", index_col=0)
conceded = pd.read_csv("data/goals_conceded_2023.csv", index_col=0)
```

```python
teams = sorted(scored.index)
teams
```

```python
BLUE = "#1d2d53"
RED = "#d22030"

plt.clf()
plt.rcParams["figure.figsize"] = (12, 8)
plt.figure(dpi=200, facecolor=Colors.WHITE)
fig, axs = plt.subplots(6, 4)
fig.patch.set_facecolor(Colors.WHITE)

for i, team in enumerate(teams):
    ax0 = axs[math.floor(i / 4) * 2][i % 4]
    ax0.set_facecolor(Colors.WHITE)
    ax0.set_title(team)
    ax0.bar(scored.columns, scored.loc[team], color=BLUE)
    ax0.set_ylim([0, 8])
    
    ax1 = axs[math.floor(i / 4) * 2 + 1][i % 4]
    ax1.set_facecolor(Colors.WHITE)
    ax1.bar(conceded.columns, conceded.loc[team], color=RED)
    ax1.set_ylim([8, 0])
    
fig.tight_layout()
```
