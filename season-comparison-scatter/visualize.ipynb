```python
import math
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from common import Fonts, Images, NWSL, Colors, add_twitter_handle, Data
```

```python
df = pd.read_csv("5_game_points_to_rank.csv")
df
```

```python
# Set up plot
plt.rcParams["figure.figsize"] = (16, 10)
plt.figure(dpi=400)
fig, ax = plt.subplots()
fig.patch.set_facecolor(Colors.WHITE)
ax.set_facecolor(Colors.WHITE)

# Plot data
ax.scatter(df["early_points"], df["rank"], c=list(map(NWSL.color, df["team"])))

# Axes
ax.set_yticks([0.08, 1])
ax.set_yticklabels(["Last", "1st"])
ax.set_xticks([0, 5, 10, 15])
ax.spines[["top", "right"]].set_visible(False)
ax.tick_params(bottom=False, left=False)
ax.tick_params(axis="both", which="major", labelsize=16, labelcolor=Colors.BLACK)
for label in ax.get_yticklabels() + ax.get_xticklabels():
    fp = Fonts.BARLOW
    fp.set_size(16)
    label.set_fontproperties(fp)

# Axis labels
ax.set_xlabel(
    "Points through 5 games",
    fontproperties=Fonts.BARLOW,
    color=Colors.BLACK,
    size=20,
)
ax.set_ylabel(
    "Final league standing",
    fontproperties=Fonts.BARLOW,
    color=Colors.BLACK,
    size=20,
)

# Title
ax.text(
    -0.75,
    1.17,
    "Early Points vs Final Standings",
    fontproperties=Fonts.BARLOW_BOLD,
    color=Colors.BLACK,
    size=28,
    ha="left",
)

# Subtitle
fig.text(
    0,
    1.03,
    "This represents the correlation between points earned through the first 5 games and final standings. No team with\n12+ points has finished in the bottom half and only one team with less than 4 points finished in the top half.",
    fontproperties=Fonts.BARLOW,
    size=18,
    transform=ax.transAxes,
    ha="left",
    color=Colors.BLACK,
)

# Twitter logo and username
add_twitter_handle(fig, x=0.8, y=0.04, fs=14, w=0.025)

# Save image
plt.savefig("5_game_points_to_rank.png", bbox_inches="tight", pad_inches=0.5, dpi=400)
```
