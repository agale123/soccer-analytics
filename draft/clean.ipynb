```python
import pandas as pd
```

```python
draft = pd.read_csv("2022_draft_registrants.csv")
draft
```

```python
draft["name"] = draft["First Name"] + " " + draft["Last Name"]
draft
```

```python
minutes  = pd.read_csv("2022_minutes.csv")
minutes
```

```python
mapping = [
        ["Savannah Marie DeMelo", "Savannah DeMelo"],
        ["Diana Ordóñez", "Diana Ordoñez"],
        ["Miri Taylor", "Miriael Taylor"],
        ["Isabel Rodriguez", "Izzy Rodriguez"]]
    
for old, new in mapping:
    minutes = minutes.replace(old, new)
```

```python
df = minutes.set_index("name").join(draft.set_index("name"), how="inner").reset_index()
df
```

```python
df.to_csv("2022_joined.csv", index=False)
```
