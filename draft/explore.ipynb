```python
import pandas as pd
from common import Data
```

```python
df = Data.get_nwsl_player_stats(include_early=True)
df = df.set_index("player").join(
    Data.get_nwsl_player_schools().set_index("name")
).reset_index()
df = df[df["school"] != "International"]
df
```

```python
df.groupby("school").agg(mins=("mins_played", "sum")).sort_values(
    "mins", ascending=False
).iloc[0:12].index
```

```python
for season in range(2013, 2024):
    if season == 2020:
        continue

    print(str(season))
    print(
        df[df["season"] == str(season)]
        .groupby("school")
        .agg(mins=("mins_played", "sum"))
        .sort_values("mins", ascending=False)
        .iloc[0:5].index
    )
```
